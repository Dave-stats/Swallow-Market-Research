```python
import pandas as pd
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSqSc00NWyTDRHxenbnnHeLmVKP3iTqPmYGqs9j2_meVGZob0nSC5eafA6EhI-JoSm1gBMN5_BNQAQ4/pub?gid=474861857&single=true&output=csv"

df = pd.read_csv(url)

print(f"Loaded  {len(df)} responses!")
df.head()
```

```python
# We remove the extra space in column names
df.columns = df.columns.str.strip()


# We want to display the column names directly now
print ("Column names in your data:")
print(df.columns.tolist())
# A quick overview of the data
print("\nData shape (rows, columns):", df.shape)
print("nFirst 3 responses:")
df.head(3)
```

```python
# We now create the most important helper columns that will help us in our analysis as we go on ahead. They are created below
df['Origin_south_east'] = df['Which region are you from?'] == 'South-East Nigeria'
df['Origin_south_west'] = df['Which region are you from?'] == 'South-West Nigerua'
df['Origin_south_south']= df['Which region are you from?'] == 'South-South Nigeria'
df['Origin_north_central']= df['Which region are you from?'] == 'North-Central Nigeria'
df['Origin_north_west'] = df['Which region are you from?'] == 'North-West Nigeria'
df['Origin_north_east'] = df['Which region are you from?'] == 'North-East Nigeria'

# Some core product flags which include our top niche demand, those that have fufu as number one, then those that have pounded yam
# as number one. Since pounded Yam is our main competitor, it should not do better than fufu in all regions else it is a problem
df['fufu_no1'] = df['What is your absolute #1 favorite swallow'] == 'Fufu'
df['Pounded_Yam_no1'] = df ['What is your absolute #1 favorite swallow'] == 'Pounded Yam'

# Frequency
#frequency_col = 'Do you eat swallows(Eba/Garri, Fufu/Akpu, Wheat, Oatmeal, Pounded Yam, Amala, ...)'
df['frequent'] = df['Do you eat swallows(Eba/Garri, Fufu/Akpu, Wheat, Oatmeal, Pounded Yam, Amala, ...etc)'].isin(['Daily', 'A few times a week'])

#Migration
df['migrant'] = df['Which region are you from?'] != df['Where do you currently live?']

#High urgency(branched only)
urgency_cond = df['Would you like your number one swallow to be available in premium format?( easy pour granules or powder ready in or under 15 mins, no pounding or long preparation)'] == 'Very interested'
optimism_cond = df['How optimistic are you that a high quality instant version could taste close to the traditionally made one?'].isin(['Very optimistic-I would buy regularly if it tastes good', 'Somewhat optimistic'])

df['high_urgency'] = urgency_cond & optimism_cond



```

```python
# STEP 8: SUPPRESSED DEMAND TRIAD AND SUBSTITUTION ANALYSIS

#1. Core Triad(Frequent + High Urgency)
triad = df[df['frequent']& df['high_urgency']]

print("===SUPPRESSED DEMAND TRIAD")
print(f"Triad size: {len(triad)} people({len(triad)/len(df): 1%} of total)\n")

print("Triad by origin zone:")
print(triad['Which region are you from?'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%\n')


# 2. Desperate but Skeptical (Very interested +low optimism)
urgency_cond = df['Would you like your number one swallow to be available in premium format?( easy pour granules or powder ready in or under 15 mins, no pounding or long preparation)'] == 'Very interested'
optimism_cond = df['How optimistic are you that a high quality instant version could taste close to the traditionally made one?'].isin(['Very optimistic-I would buy regularly if it tastes good', 'Somewhat optimistic'])
neutral_cond = df['How optimistic are you that a high quality instant version could taste close to the traditionally made one?'] == 'Neutral'
skepticism_cond= df['How optimistic are you that a high quality instant version could taste close to the traditionally made one?'].isin(['Not sure', 'Not optimistic-doubt it can match traditional'])

# flags

df['high_urgency'] = urgency_cond & optimism_cond
df['indifferent'] = urgency_cond & neutral_cond
df['desperate_but_skeptical'] = urgency_cond & skepticism_cond


# Summary
print(f"High Urgency: {df['high_urgency'].sum()}")
print(f"Desperate but Skeptical: {df['desperate_but_skeptical'].sum()}")
print(f"Neutral): {df['desperate_but_skeptical'].sum()}")

skeptical = df[df['desperate_but_skeptical']]
print(f"size: {len(skeptical)} people({len(skeptical)/len(df):.1%} of total)\n")
print("Their #1 swallows:")
print(skeptical['What is your absolute #1 favorite swallow'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%\n')
print(f"% frequent among desperate skeptical: {skeptical['frequent'].mean():.1%}\n")

# 4 Flipped suppresion signal(for triad swallows)
category_freq = df['frequent'].mean()*100
print("Flipped frequency loyalty for triad #1 swallows (P(Frequent|#1 Swallow))")
for swallow in triad['What is your absolute #1 favorite swallow'].unique():
    loyalty = df[df['What is your absolute #1 favorite swallow']== swallow]['frequent'].mean()*100
    print(f"{swallow}: {loyalty:.1f}% (category average: {category_freq:.1f}%)")
    
```

```python
# Step 11: GRID-BASED LATENT AND SUBSTITUTION DEMAND

# 1. Exact Column Headers from your screenshot
fufu_col = "Rank your top 3 favorite swallows [Fufu/Akpu]"
yam_col = "Rank your top 3 favorite swallows [Pounded Yam]"

# 2. Function to catch "1st", "Ist", "2nd", etc.
def get_rank(value):
    val = str(value).strip().lower()
    if 'ist' in val or '1st' in val: return 1
    if '2nd' in val: return 2
    if '3rd' in val: return 3
    return 0

# 3. Apply the mapping
df['fufu_rank'] = df[fufu_col].apply(get_rank)
df['yam_rank'] = df[yam_col].apply(get_rank)

# 4. Create logical flags for analysis
df['fufu_no1'] = df['fufu_rank'] == 1
df['fufu_in_top2'] = (df['fufu_rank'] > 0) & (df['fufu_rank'] <= 2)
df['fufu_in_top3'] = (df['fufu_rank'] > 0) & (df['fufu_rank'] <= 3)
df['pounded_yam_no1'] = df['yam_rank'] == 1

# 5. Output Results
print("=== MARKET URGENCY: FUFU LATENT DEMAND ===")
print(f"Fufu Loyalists (#1): {df['fufu_no1'].mean():.1%}")
print(f"Fufu Top 2 (Strong latent demand): {df['fufu_in_top2'].mean():.1%}")
print(f"Fufu Top 3 (Total reachable market): {df['fufu_in_top3'].mean():.1%}")

# 6. PRIME SWITCHERS (Targeting Pounded Yam lovers who accept Fufu)
# They picked Yam as #1, but put Fufu as 2nd or 3rd
prime_switchers = df[df['pounded_yam_no1'] & (df['fufu_rank'] > 1)]

print(f"\n--- PRIME SWITCHER ANALYSIS ---")
print(f"Number of Prime Switchers: {len(prime_switchers)}")
print(f"Market Share of Switchers: {len(prime_switchers)/len(df):.1%}")
```
